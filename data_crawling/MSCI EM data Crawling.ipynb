```python
from urllib.request import urlopen
import json
import pandas as pd
```

```python
url = 'https://www.msci.com/documents/1296102/19200117/em_weight.json/8ccc1d87-993b-2391-7f87-1baf55b86478'
```

```python
json_obj = json.load(html)
```

```python
print(json_obj)
```
